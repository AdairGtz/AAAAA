```python
X1= X.iloc[:, :]
y2=y.iloc[:]
```

```python
#capa = tf.keras.layers.Dense(units=138, input_shape=[13])
#modelo = tf.keras.Sequential([capa])

oculta1 = tf.keras.layers.Dense(units=84, input_shape=[13], activation='relu')
oculta2 = tf.keras.layers.Dense(units=168, activation='relu')
salida = tf.keras.layers.Dense(units=1)

modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    loss='mean_squared_error'
)
```

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(X1, y2, epochs=1082, verbose=False)
print("Modelo entrenado!")
```

```python
print("Hagamos una prediccin!")
resultado = modelo.predict([X.iloc[[0], :]])
#resultado = modelo.predict([[0.789474,	0.776596,	0.8125	,0.692308,	0.000000,	0.363636,	0.884211,	0.6320,	0.384615,	0.000000,	1.000000,	0.137931,	0.583333]])
print("El resultado es " + str(resultado) + " USD!")
```
